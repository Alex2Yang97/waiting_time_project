```python
import pymysql
import pandas as pd
import numpy as np
```

```python
def get_df_from_sql(sql):
    # Connect to the MySQL database
    db = pymysql.connect(host='localhost',
                         port=3306,
                         user='root',
                         passwd='root',
                         db='originaldata')
    df = pd.read_sql(sql = sql, con = db)
    return df
```

```python
sql_alias = """SELECT * FROM alias"""
sql_PL = """SELECT * FROM Patientlocation"""
sql_PLMH = """SELECT * FROM Patientlocationmh"""
sql_appointment = """
    SELECT * FROM Appointment ap
    WHERE
        ap.ActualStartDate NOT IN ('0000-00-00 00:00:00','1970-01-01 00:00:00')
    AND ap.ActualEndDate NOT IN ('0000-00-00 00:00:00','1970-01-01 00:00:00')
    AND ap.ScheduledStartTime NOT IN ('0000-00-00 00:00:00','1970-01-01 00:00:00')
    AND ap.ScheduledEndTime NOT IN ('0000-00-00 00:00:00','1970-01-01 00:00:00')
    AND ap.ActualEndDate NOT IN ('0000-00-00 00:00:00','1970-01-01 00:00:00')"""

data_alias = get_df_from_sql(sql_alias)
data_PL = get_df_from_sql(sql_PL)
data_PLMH = get_df_from_sql(sql_PLMH)
data_appointment = get_df_from_sql(sql_appointment)
```
