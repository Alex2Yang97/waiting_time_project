```python
# 保存预处理之后的数据，pickle文件
def pickle_dump(data_to_pickle, filename, local_path = local_path):
    """ Function to write data to pickle format. """
    pickle.dump(data_to_pickle, open(local_path + filename, "wb+"))
```

```python
# 读取预处理之后的数据，pickle文件

def pickle_load(filename, local_path = local_path):
    with open(local_path + filename, 'rb') as f:
        u = pickle._Unpickler(f)
        u.encoding = 'latin1'
        data = u.load()
    return data
```

```python
print('开始-得到所有统计量！')
# 数量过多的的特征
PatientSerNum = []
AppointmentSerNum = []
allocated = []
CourseSerNum = []
DoctorSerNum = []
PlanSerNum = []
duration = []

# 全部数据的统计量
DiagnosisName_list = []
Sex_list = []
age_list = []
AliasSerNum_list = []
orientation_list = []
PrimaryFlag_list = []
OncologistFlag_list = []
Status_list = []
CourseId_list = []
AliasName_list = []
AliasType_list = []

# 时长相关的统计量
total_duration_list = []
check_call_duration_list = []
call_actual_duration_list = []
actual_scheduled_duration_list = []
# 一次治疗的实际时长
actual_duration_list = []
# 一次治疗的预期时长 allocated_duration
scheduled_duration_list = []
start_radiation_duration_list = []
radiation_end_duration_list = []

# radiation 相关的特征
ImagesTaken_list = []
ImageDuration_list = []
UserName_list = []
RadiationId_list = []
MU_list = []
MUCoeff_list = []
CourseId_list = []


appointment_num = 0
course_num = 0
for PatientSerNum in raw_patients.keys():
    
    # 完整治疗时长
    appointment_sequence = sorted(list(raw_patients[PatientSerNum].keys()))
    index = appointment_sequence[0]
    CompletedDateTime = raw_patients[PatientSerNum][index]['CompletedDateTime'][0]
    StartDateTime = raw_patients[PatientSerNum][index]['StartDateTime'][0]
    total_duration = (CompletedDateTime - StartDateTime).total_seconds()
    
    if total_duration <= 0:
        index = appointment_sequence[-1]
        CompletedDateTime = raw_patients[PatientSerNum][index]['StartDateTime'][0]
        total_duration = (CompletedDateTime - StartDateTime).total_seconds()
    total_duration_list.append(total_duration)
    
    for Scheduled_Start_Time in sorted(raw_patients[PatientSerNum].keys()):
        
        # 统计治疗的总数
        appointment_num = appointment_num + 1
        if appointment_num % 1000 == 0:
            print('The Number of Appointment is ', appointment_num)
        
        # 不同特征的统计量
        appointment = raw_patients[PatientSerNum][Scheduled_Start_Time]
        DiagnosisName = appointment['DiagnosisName'][0]
        Sex = appointment['Sex'][0]
        age = appointment['age'][0]
        AliasSerNum = appointment['AliasSerNum'][0]
        orientation = appointment['orientation'][0]
        PrimaryFlag = appointment['PrimaryFlag'][0]
        OncologistFlag = appointment['OncologistFlag'][0]
        Status = appointment['Status'][0]
        CourseId = appointment['CourseId'][0]
        AliasName = appointment['AliasName'][0]
        AliasType = appointment['AliasType'][0]
        
        DiagnosisName_list.append(DiagnosisName)
        Sex_list.append(Sex)
        age_list.append(age)
        AliasSerNum_list.append(AliasSerNum)
        orientation_list.append(orientation)
        PrimaryFlag_list.append(PrimaryFlag)
        OncologistFlag_list.append(OncologistFlag)
        Status_list.append(Status)
        CourseId_list.append(CourseId)
        AliasName_list.append(AliasName)
        AliasType_list.append(AliasType)
        
        # 时间段的统计量
        ScheduledStartTime = appointment['ScheduledStartTime'][0]
        ScheduledEndTime = appointment['ScheduledEndTime'][0]
        ActualStartDate = appointment['ActualStartDate'][0]
        ActualEndDate = appointment['ActualEndDate'][0]
        
        # check in
        ArrivalDateTimeh = appointment['ArrivalDateTimeh'][0]
        #HstryDateTime = appointment['HstryDateTime']
        
        # call in
        ArrivalDateTime = appointment['ArrivalDateTime'][0]
        
        # 一次治疗check in 和 call in 之间的时间间隔
        check_call_duration = (ArrivalDateTime - ArrivalDateTimeh).total_seconds()
        check_call_duration_list.append(check_call_duration)
        
        # 一次治疗actual start 和 call in 之间的时间间隔
        call_actual_duration = (ActualStartDate - ArrivalDateTime).total_seconds()
        call_actual_duration_list.append(call_actual_duration)
        
        # 预期的开始时间与真实开始时间之间的时间间隔
        actual_scheduled_duration = abs((ActualStartDate - ArrivalDateTime).total_seconds())
        actual_scheduled_duration_list.append(actual_scheduled_duration)
        
        # 一次治疗的实际时长
        actual_duration = (ActualEndDate - ActualStartDate).total_seconds()
        actual_duration_list.append(actual_duration)
        
        # 一次治疗的预期时长 allocated_duration
        scheduled_duration = (ScheduledEndTime - ScheduledStartTime).total_seconds()
        scheduled_duration_list.append(scheduled_duration)
        
        # 对于含有course_information的进行单独统计
        if 'course_information' in appointment.keys():
            course_num = course_num + 1
            course_information = appointment['course_information']
            
            for RadiationSerNum in course_information.keys():
                
                # 统计接受 radiation的数量
                course_num = course_num + 1
                
                # 统计radiation的特征
                course = course_information[RadiationSerNum]
                ImagesTaken = course['ImagesTaken'][0]
                ImageDuration = course['ImageDuration'][0]
                UserName = course['UserName'][0]
                RadiationId = course['RadiationId'][0]
                MU = course['MU'][0]
                MUCoeff = course['MUCoeff'][0]
                CourseId = course['CourseId'][0]
                tx_start = course['tx_start'][0]
                tx_end = course['tx_end'][0]
                
                ImagesTaken_list.append(ImagesTaken)
                ImageDuration_list.append(ImageDuration)
                UserName_list.append(UserName)
                RadiationId_list.append(RadiationId)
                MU_list.append(MU)
                MUCoeff_list.append(MUCoeff)
                CourseId_list.append(CourseId)
                
                # 从开始治疗 actual start到 radiation开始的时长
                start_radiation_duration = (tx_start - ActualStartDate).total_seconds()
                start_radiation_duration_list.append(start_radiation_duration)
                
                # 从结束 radiation到actual end转变
                radiation_end_duration = (ActualEndDate - tx_end).total_seconds()
                radiation_end_duration_list.append(radiation_end_duration)

print('完成-得到所有统计量！')
```

```python
def getMeanAndStd(duration_list):
    mean = np.mean(duration_list)
    std = np.std(duration_list)
    return mean, std
```
