```python
def get_df_from_sql(sql):
    # Connect to the MySQL database
    db = pymysql.connect(host='localhost',
                         port=3306,
                         user='root',
                         passwd='root',
                         db='originaldata')
    df = pd.read_sql(sql = sql, con = db)
    return df
```

```python
# 计划治疗时长
data['interval_ScheduledStartTime_ScheduledEndTime'] = data['ScheduledEndTime'] - data['ScheduledStartTime']
data['interval_ScheduledStartTime_ScheduledEndTime'] = data['interval_ScheduledStartTime_ScheduledEndTime'].apply(
    lambda x: x.total_seconds())
# 实际治疗时长
data['interval_ActualStartDate_ActualEndDate'] = data['ActualEndDate'] - data['ActualStartDate']
data['interval_ActualStartDate_ActualEndDate'] = data['interval_ActualStartDate_ActualEndDate'].apply(
    lambda x: x.total_seconds())
# 真实治疗开始时间--和--计划治疗开始时间--的间隔
data['interval_ActualStartDate_ScheduledStartTime'] = data['ScheduledStartTime'] - data['ActualStartDate']
data['interval_ActualStartDate_ScheduledStartTime'] = data['interval_ActualStartDate_ScheduledStartTime'].apply(
    lambda x: x.total_seconds())
# 真实治疗结束时间-和--计划治疗结束时间---的间隔
data['interval_ActualEndDate_ScheduledEndTime'] = data['ScheduledEndTime'] - data['ActualEndDate']
data['interval_ActualEndDate_ScheduledEndTime'] = data['interval_ActualEndDate_ScheduledEndTime'].apply(
    lambda x: x.total_seconds())

# 同一个治疗中，有多个ArrivalDateTime
# 每个ArrivalDateTime--和--ActualStartDate--的间隔
data['interval_ArrivalDateTime_ActualStartDate'] = data['ActualStartDate'] - data['ArrivalDateTime']
data['interval_ArrivalDateTime_ActualStartDate'] = data['interval_ArrivalDateTime_ActualStartDate'].apply(
    lambda x: x.total_seconds())
# 每个HstryDateTime--和--ActualStartDate--的间隔
data['interval_HstryDateTime_ActualStartDate'] = data['ActualStartDate'] - data['HstryDateTime']
data['interval_HstryDateTime_ActualStartDate'] = data['interval_HstryDateTime_ActualStartDate'].apply(
    lambda x: x.total_seconds())
# 每个pltArrivalDateTime--和--ActualStartDate--的间隔
data['interval_pltArrivalDateTime_ActualStartDate'] = data['ActualStartDate'] - data['pltArrivalDateTime']
data['interval_pltArrivalDateTime_ActualStartDate'] = data['interval_pltArrivalDateTime_ActualStartDate'].apply(
    lambda x: x.total_seconds())

# 每个ArrivalDateTime--和--ActualEndDate--的间隔
data['interval_ArrivalDateTime_ActualEndDate'] = data['ActualEndDate'] - data['ArrivalDateTime']
data['interval_ArrivalDateTime_ActualEndDate'] = data['interval_ArrivalDateTime_ActualEndDate'].apply(
    lambda x: x.total_seconds())
# 每个HstryDateTime--和--ActualEndDate--的间隔
data['interval_HstryDateTime_ActualEndDate'] = data['ActualEndDate'] - data['HstryDateTime']
data['interval_HstryDateTime_ActualEndDate'] = data['interval_HstryDateTime_ActualEndDate'].apply(
    lambda x: x.total_seconds())
# 每个pltArrivalDateTime--和--ActualEndDate--的间隔
data['interval_pltArrivalDateTime_ActualEndDate'] = data['ActualEndDate'] - data['pltArrivalDateTime']
data['interval_pltArrivalDateTime_ActualEndDate'] = data['interval_pltArrivalDateTime_ActualEndDate'].apply(
    lambda x: x.total_seconds())

# 每个ArrivalDateTime--和--HstryDateTime--的间隔
data['interval_ArrivalDateTime_HstryDateTime'] = data['HstryDateTime'] - data['ArrivalDateTime']
data['interval_ArrivalDateTime_HstryDateTime'] = data['interval_ArrivalDateTime_HstryDateTime'].apply(
    lambda x: x.total_seconds())
# 每个pltArrivalDateTime--和--HstryDateTime--的间隔
data['interval_pltArrivalDateTime_HstryDateTime'] = data['HstryDateTime'] - data['pltArrivalDateTime']
data['interval_pltArrivalDateTime_HstryDateTime'] = data['interval_pltArrivalDateTime_HstryDateTime'].apply(
    lambda x: x.total_seconds())
# 每个pltArrivalDateTime--和--ArrivalDateTime--的间隔
data['interval_ArrivalDateTime_pltArrivalDateTime'] = data['pltArrivalDateTime'] - data['ArrivalDateTime']
data['interval_ArrivalDateTime_pltArrivalDateTime'] = data['interval_ArrivalDateTime_pltArrivalDateTime'].apply(
    lambda x: x.total_seconds())
```

```python
# 删除异常样本
def get_normal_interval(column_name, data):
    normal_interval = data[['PatientSerNum', 'AppointmentSerNum', column_name]].copy()
    normal_interval = normal_interval[(normal_interval[column_name] <= 60*60*12)&
                                      (normal_interval[column_name] >= -60*60*12)]
    return normal_interval
```

```python
interval_grouped = interval.groupby('AppointmentSerNum')
interval_appt = interval_grouped.mean()
interval_appt = interval_appt.sort_values(by = column_name)
normal_number = interval_appt[(interval_appt[column_name]) <= 60*15].shape[0]
print(f'The number of normal sample {normal_number}')
# y = mean['interval_asd_aed'].tolist()

std = interval_appt.describe()[column_name].loc['std']
print(f'Standard deviation is {std}')
mean = interval_appt.describe()[column_name].loc['mean']
print(f'Mean is {mean}')

shangjie = interval_appt[(interval_appt[column_name]) <= mean - 3*std].shape[0]
xiajie = interval_appt[(interval_appt[column_name]) <= mean + 3*std].shape[0]
```

```python
y = interval_appt[column_name].tolist()
x = range(len(y))

fig = plt.figure(figsize=(10, 10), dpi = 300)
plt.plot(x, y, c = 'blue', label = 'interval distribution')
plt.plot(normal_number, 60*15, 'ks')
# 分割线
# plt.plot(x, [60*15]*len(y), 'r--')
plt.axvline(normal_number, linestyle = '--', label = 'normal interval')
plt.axvline(shangjie, linestyle = '--', c = 'black', label = '-3 Standard deviation')
plt.axvline(xiajie, linestyle = '--', c = 'black', label = '+3 Standard deviation')
# 横线
plt.axhline(mean, linestyle = '-.', c = 'r', label = 'mean')
# 标注特殊点
plt.annotate(f'normal sample: {normal_number}',
             xytext = (normal_number-10000, 60*15+1000), xy = (normal_number, 60*15), fontsize = 10)
plt.annotate(f'+3 std sample: {xiajie}',
             xytext = (xiajie-10000, mean+3*std+1000), xy = (xiajie-10000, mean-3*std), fontsize = 10)
plt.annotate(f'-3 std sample: {shangjie}',
             xytext = (shangjie-10000, mean+3*std+1000), xy = (shangjie-10000, mean-3*std), fontsize = 10)
plt.annotate(f'total sample: {len(x)}',
             xytext = (x[-1]-10000, y[-1]+1000), xy = (x[-1], y[-1]), fontsize = 10)
# plt.plot(x, [60*10]*len(y), c = 'r')
# plt.plot(x, [60*30]*len(y), c = 'r')
# plt.plot(x, [60*60]*len(y), c = 'r')
plt.xlabel('The number of appointments', fontsize = 20)
plt.ylabel(f'{column_name}', fontsize = 20)
plt.title(f'The distribution of {column_name}', fontsize = 20)
plt.legend(loc = 'best')
plt.savefig(picture_path + f'{column_name}.jpg')
plt.show()
```
