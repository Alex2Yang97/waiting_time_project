```python
import pickle
import MySQLdb
import numpy as np
import random
import matplotlib.pyplot as plt
import datetime
import operator as op
import math
import random
import pandas as pd
from collections import Counter
```

```python
# Connect to the MySQL database
def getSQLtest(sql):
    # Connect to the MySQL database
    db = MySQLdb.connect(host='localhost',
                         port=3306,
                         user='root',
                         passwd='root',
                         db='originaldata')
    db_cursor = db.cursor(MySQLdb.cursors.DictCursor)
    sql0 = """
         set sql_mode = 'STRICT_TRANS_TABLES,NO_ZERO_IN_DATE,NO_ZERO_DATE,ERROR_FOR_DIVISION_BY_ZERO,NO_ENGINE_SUBSTITUTION';
    """       
    db_cursor.execute(sql0)
    db_cursor.execute(sql)
    db_results = db_cursor.fetchall()
    db_cursor.close()
    print ("Fetched Preliminary query results")
    print ("Number of results: " + str(len(db_results)))
    return db_results
```

```python
# Preliminary query for feature data
sql1 = """
    SELECT * from radiation
"""
```

```python
print("Executing query...")
db_results1 = getSQLtest(sql1)
print('Finish!')
```

```python
print('Start getting raw_patients dictionary')
count = 1
raw_patients = dict()
RadiationId_list = []
TechniqueId_list = []
DeliveryType_list = []

# Loop through each result
for row in db_results1:
    RadiationId = row['RadiationId']
    TechniqueId = row['TechniqueId']
    DeliveryType = row['DeliveryType']
    
    RadiationId_list.append(RadiationId)
    TechniqueId_list.append(TechniqueId)
    DeliveryType_list.append(DeliveryType)
    
    
print('Finish getting raw_patients dictionary')
```

```python
# Preliminary query for feature data
sql2 = """
    SELECT * from resource
"""
print("Executing query...")
db_results1 = getSQLtest(sql2)
print('Finish!')
```

```python
print('Start getting raw_patients dictionary')

ResourceName_list = []
ResourceType_list = []
ResourceName_list_Auxiliary = []
ResourceName_list_Doctor = []
ResourceName_list_Machine = []
ResourceName_list_Staff = []
ResourceName_list_Venue = []

count = 0
# Loop through each result
# ResourceType: '', 'Auxiliary', 'Doctor', 'Machine', 'Staff', 'Venue'
for row in db_results1:
    #print(row)
    #break
    
    ResourceName = row['ResourceName']
    ResourceType = row['ResourceType']
    
    if ResourceType == 'Auxiliary':
        ResourceName_list_Auxiliary.append(ResourceName)
    elif ResourceType == 'Doctor':
        ResourceName_list_Doctor.append(ResourceName)
    elif ResourceType == 'Machine':
        ResourceName_list_Machine.append(ResourceName)
    elif ResourceType == 'Staff':
        ResourceName_list_Staff.append(ResourceName)
    elif ResourceType == 'Venue':
        ResourceName_list_Venue.append(ResourceName)
    else:
        count = count + 1
    
    
    ResourceName_list.append(ResourceName)
    ResourceType_list.append(ResourceType)
    
    
print('Finish getting raw_patients dictionary')
```
